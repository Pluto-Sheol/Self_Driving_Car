```python
#Chuẩn bị dữ liệu
import os
import numpy as np
#"obj" là tên thư mục chứa cả ảnh và file annotation.
lst_files = os.listdir("/content/gdrive/MyDrive/yolov4/darknet/data/images/")
lst_images = []

for file in lst_files:
  if ".txt" not in file:
    lst_images.append(file)

#Tách 10% ảnh ra làm tập validation
random_idx = np.random.randint(0, len(lst_images), int(len(lst_images)*0.1))

#Tạo file train.txt được đặt trong thư mục darknet/data
with open("/content/gdrive/MyDrive/yolov4/darknet/data/train.txt","w") as f:
  for idx in range(len(lst_images)):
    if idx not in random_idx:
      f.write("/content/gdrive/MyDrive/yolov4/darknet/data/images/"+lst_images[idx]+"\n")

#Tạo file valid.txt được đặt trong thư mục darknet/data
with open("/content/gdrive/MyDrive/yolov4/darknet/data/valid.txt","w") as f:
    for idx in random_idx:
      f.write("/content/gdrive/MyDrive/yolov4/darknet/data/images/"+lst_images[idx]+"\n")
```
